```python
# import libraries
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# reading an image with imread command
myImg = cv.imread(r"E:\GitHub\OpenCV\images\Myself.jpg")
```

```python
# show the image:
cv.imshow('Me', myImg)
```

```python
# app must run until a button or an specific action happened, so
# we must use a command named WAITKEY
cv.waitKey()
cv.destroyAllWindows()
```

```python
plt.imshow(myImg)
```

```python
# convert image to grayscale with matplotlib:
myImg2 = cv.imread(r"E:\GitHub\OpenCV\images\Myself.jpg", 0) # ---> 0 converts image to grayscale when image is on reading
cv.imshow('gray', myImg2)
plt.imshow(myImg2, cmap = 'gray')
cv.waitKey()
cv.destroyAllWindows()
```
